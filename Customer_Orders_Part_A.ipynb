```python
import pandas as pd

# Step 2: Load the dataset
df = pd.read_csv('customer_orders.csv')  # make sure the file is in the same folder or uploaded
df.head(10)  # Step 3: Display first 10 rows
```

```python
df.isnull().sum()
```

```python
# Step 5: Handle missing values
df['PurchaseAmount'].fillna(df['PurchaseAmount'].mean(), inplace=True)
df.dropna(inplace=True)  # if there are any remaining
```

```python
df['PurchaseDate'] = pd.to_datetime(df['PurchaseDate'], errors='coerce')
```

```python
df['PurchaseMonth'] = df['PurchaseDate'].dt.month
```

```python
# Step 8: Create AgeGroup column
bins = [0, 25, 35, 45, 60, 100]
labels = ['18-25', '26-35', '36-45', '46-60', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```
